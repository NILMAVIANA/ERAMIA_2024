## Ciência de dados
</div>

Com o Colab, você pode aproveitar todo o potencial das conhecidas bibliotecas Python para analisar e ver dados. A célula de códigos abaixo usa <strong>numpy</strong> para gerar dados aleatórios e <strong>matplotlib</strong> para visualizá-los. Para editar o código, basta clicar na célula e começar a editar.

É possível importar para os notebooks do Colab os dados da sua conta do Google Drive, como planilhas. Também é possível importar do GitHub e de muitas outras fontes. Para saber mais sobre como importar dados e como o Colab pode ser usado para a ciência de dados, consulte o link abaixo em <a href="#working-with-data">Como trabalhar com dados</a>.

```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```

```

import numpy as np
import matplotlib.pyplot as plt
import urllib.request
from skimage import io, color  # Substitui o PIL

# Baixar a imagem
urllib.request.urlretrieve(
    'https://blog-static.petlove.com.br/wp-content/uploads/2021/09/gato-cinza-petlove7.jpg',
    'imagem.jpg'
)

# Carregar a imagem usando skimage
imagem_RGB = io.imread('imagem.jpg')

# Mostrar a imagem
plt.figure(1)
io.imshow(imagem_RGB)  # Pode usar tanto io.imshow quanto plt.imshow
#plt.axis('off')  # Opcional: remove os eixos
#plt.show()
```

```
# Dimensão da imagem
print(imagem_RGB.shape)
```

```
# Altura
print(imagem_RGB.shape[0])
```

```
# Largura
print(imagem_RGB.shape[1])
```

```
# Carregar a imagem diretamente em escala de cinza com skimage
# imagem_cinza = io.imread('imagem.bmp', as_gray=True)
imagem_cinza = (io.imread('imagem.jpg', as_gray=True) * 255).astype('uint8')

# Mostrar a imagem
plt.figure(1)
plt.imshow(imagem_cinza, cmap='gray')
plt.axis('off')  # Remove os eixos
plt.show()
```

```

# Converter a imagem para um array numpy
matriz = np.array(imagem_cinza)

h = np.zeros(256)
for i in range(matriz.shape[0]):
  for j in range(matriz.shape[1]):
    h[matriz[i,j]] += 1

# Plotar o gráfico
eixo_x = list(range(256))
plt.figure(2)
plt.bar(eixo_x, h)
plt.show()
```
